```python
import pandas as pd #data manipulation lib
import matplotlib.pyplot as plt #data visualization lib
import seaborn as sns
mercedes_df=pd.read_csv('mercedes.csv')
mercedes_df
```

```python
mercedes_df.isnull().sum()
missing_values_df=pd.DataFrame({'col_name':mercedes_df.columns, 
                                'missing_values_count':mercedes_df.isnull().sum(),
                               'missing_value_percentage':mercedes_df.isnull().sum()/mercedes_df.shape[0]*100})
mercedes_df
```

```python
cols_to_drop=[]


column_to_drop=list(missing_values_df[missing_values_df['missing_value_percentage']>50].index)

for col in mercedes_df.columns:
    if mercedes_df[col].nunique()==mercedes_df.shape[0] or mercedes_df[col].nunique()==1:
        cols_to_drop.append(col)

abc=column_to_drop+cols_to_drop

mercedes_df.drop(columns=abc,inplace=True)
mercedes_df
```

```python
for col in mercedes_df:
    mercedes_df[col]=mercedes_df[col].astype(object)
mercedes_df.dtypes
```

```python
y=mercedes_df['y']
mercedes_df.drop(columns='y',inplace=True)
```

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(mercedes_df,y,test_size=0.2,random_state=43)  #y test and y train are outputs

#filling missing values


for col in x_train.columns:
    if x_train[col].dtype=='object':
        x_train[col]=x_train[col].fillna(x_train[col].mode()[0]).astype(object)
        x_test[col]=x_test[col].fillna(x_train[col].mode()[0]).astype(object)
    else:
        x_train[col]=x_train[col].fillna(x_train[col].mean())
        x_test[col]=x_test[col].fillna(x_train[col].mean())
        
x_train.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoderExt()


for col in x_train.columns:
    if x_train[col].dtype=='object':
        le.fit(x_train[col])
        x_train[col]=le.transform(x_train[col])
        x_test[col]=le.transform(x_test[col])
        
x_train[col]
```

```python
from sklearn.decomposition import PCA
pca=PCA()
x_train_pca_df=pd.DataFrame(pca.fit_transform(x_train))
x_train_pca_df
pca.explained_variance_ratio_
#plt.plot(np.cumsum(pca.explained_variance_ratio_))
# plt.xlabel()
```

```python
pca_1=PCA(n_components=50)

x_train_pca_df=pd.DataFrame(pca_1.fit_transform(x_train),columns=list(range(0,50)))
x_test_pca_df=pd.DataFrame(pca_1.transform(x_test))
x_train_pca_df
```

```python
# For each X, calculate VIF and save in dataframe
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif = pd.DataFrame()
vif["VIF_Factor"] = [variance_inflation_factor(x_train.values, i) for i in range(x_train.shape[1])]
vif['column_names'] = x_train.columns
vif
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_pca = pd.DataFrame()
vif_pca["VIF_Factor_PCA"] = [variance_inflation_factor(x_train_pca_df.values, i) for i in range(x_train_pca_df.shape[1])]
vif_pca['column_names'] = x_train_pca_df.columns
vif_pca
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_squared_error


linreg=LinearRegression()

linreg.fit(x_train_pca_df,y_train)

y_pred=linreg.predict(x_test_pca_df)
y_pred


print('R2',r2_score(y_test,y_pred))
print('R2 linear reg class',linreg.score(x_test_pca_df,y_test))
linreg.coef_
```
